## **Common Password Dataset**

**Source:** 

1. [Dataset 1 - Kaggle](https://www.kaggle.com/datasets/shivamb/10000-most-common-passwords)
2. [Dataset 2 - Kaggle](https://www.kaggle.com/datasets/wjburns/common-password-list-rockyoutxt)
3. [Dataset 3 - Kaggle](https://www.kaggle.com/datasets/bhavikbb/password-strength-classifier-dataset)
4. [Dataset 4 - Kaggle](https://www.kaggle.com/datasets/utkarshx27/passwords)

The following codes are used to produce a **clean dataset** that will later be used as a **list of common passwords.**

```python
# Import the necessary libraries
import pandas as pd
```

```python
# Read dataset (.csv)
df1 = pd.read_csv("datasets/data1.csv", header = 0, sep = ",")

df1.head()
```

```python
# Create a DataFrame
df1 = df1["password"]
df1 = pd.DataFrame(df1)

df1.head()
```

```python
# Perform data preprocessing
df1 = df1.dropna()
df1 = df1.drop_duplicates()

df1.head()
```

```python
# Read dataset (.csv)
df2 = pd.read_csv("datasets/data2.csv", header = 0, sep = ",")

df2.head()
```

```python
# Create a DataFrame
df2 = df2["password"]
df2 = pd.DataFrame(df2)

df2.head()
```

```python
# Perform data preprocessing
df2 = df2.dropna()
df2 = df2.drop_duplicates()

df2.head()
```

```python
# Read dataset (.csv)
df3 = pd.read_csv("datasets/data3.csv", header = 0, sep = ",")

df3.head()
```

```python
# Create a DataFrame
df3 = df3["password"]
df3 = pd.DataFrame(df3)

df3.head()
```

```python
# Perform data preprocessing
df3 = df3.dropna()
df3 = df3.drop_duplicates()

df3.head()
```

```python
# Read dataset (.csv)
df4 = pd.read_csv("datasets/data4.csv", header = 0, sep = ",")

df4.head()
```

```python
# Perform data preprocessing
df4 = df4.dropna()
df4 = df4.drop_duplicates()

df4.head()
```

```python
# Combining dataframes
df = [df1, df2, df3, df4]
df = pd.concat(df, ignore_index = True)

df.head()
```

```python
# Perform data preprocessing and sort the data
df = df.dropna()
df = df.drop_duplicates()
df = df.sort_values(by = "password")
```

```python
# Save the data to a csv file
df.to_csv("dataset.txt", index = False, sep = "\t")
```
